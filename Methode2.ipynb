```python
#ici, ggplot2 est très pratique mais les données doivent être formatées dans un format 'long'.
#Le format que nous avions auparavant est appelé 'wide' (parfois, format 'table').
%%R
library(reshape)
iris_melt <- reshape::melt(Sver)
#les facettes sont des parcelles multi-plan. Nous pouvons visualiser les données par fonction, classe..
g <- ggplot(iris_melt, aes(x=value, fill=variable)) + geom_histogram(bins=10, color='white')
g <- g + facet_wrap(.~variable)
g
```

```python
#garder seulement la fonctionnalité qui nous intéresse, ici 'sepal.width'
%%R
library(reshape)
iris_melt <- reshape::melt(Sver)

SverW_sepal_feature <- iris_melt[iris_melt[,'variable']=='sepal_width',]
options(repr.plot.width=8, repr.plot.height=6)
g <- ggplot(SverW_sepal_feature, aes(x=value, fill=species))
g <- g + geom_histogram(bins=20, color='white', position='identity', alpha=0.75)
g <- g + xlab('width (cm)')
g <- g + ggtitle ('Distribution de sepal_width des classes setosa et versicolor')
g
```

```python
%%R
options(repr.plot.width=8, repr.plot.height=6)

g <- ggplot(iris, aes(x=petal_length, y=petal_width, color=species)) + geom_point(size=5)
g <- g + ggtitle('iris : Petal_length vs Petal_width')
g <- g + xlab('Petal length (cm)')
g <- g + ylab('Petal width (cm)')
g <- g + theme_bw() 
#+ font_change
g
```

```python
#la fonction 'grep()' vérifie si les termes longueur ou largeur sont présents dans les noms de colonnes
#Si l’un des termes est présent, l’index de la colonne est retourné
#la fonction 'colnames()' retourne les noms des colonnes
%%R
index_col <- grep('length|width', colnames(iris), perl=TRUE)
irisACP <- iris[,index_col]
#prcomp fonction de l'ACP
cp <- prcomp(irisACP, center=T, scale=T)
cp
```

```python
%%R
summary(cp)
```

```python
%%R
barplot(summary(cp)$ importance[2,], ylab='Proportion of variance explained')
```

```python
%%R
library(cowplot)
options(repr.plot.width=12, repr.plot.height=4)

acp <- cbind(as.data.frame(cp$x), species=iris$species)
g0 <- ggplot(acp, aes(x=PC1, y=PC2, color=species)) + geom_point(size=4, alpha=0.6)
g0 <- g0 + theme_bw()

g1 <- ggplot(acp, aes(x=PC1, y=PC3, color=species)) + geom_point(size=4, alpha=0.6)
g1 <- g1 + theme_bw()

plot_grid(g0, g1, labels="AUTO")
```
